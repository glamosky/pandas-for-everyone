```python
import seaborn as sns
anscombe = sns.load_dataset("anscombe")
print(anscombe)
```

```python
import matplotlib.pyplot as plt
```

```python
#create a subset of the data
dataset_1 = anscombe[anscombe['dataset'] == 'I']
```

```python
plt.plot(dataset_1['x'], dataset_1['y'],)
plt.show()
```

```python
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']
```

```python
#create the entire figure where our subplots will go
fig = plt.figure()

#tell the figure how the subplots should be arranged
#in the example below, we have 2 rows of plots and each row will have 2 plots

#subplot has 2 rows and 2 columns, plot location 1
axes1 = fig.add_subplot(2, 2, 1)

#subplot has 2 rows and 2 columns, plot location 2
axes2 = fig.add_subplot(2, 2, 2)

#subplot has 2 rows and 2 columns, plot location 3
axes3 = fig.add_subplot(2, 2, 3)

#subplot has 2 rows and 2 columns, plot location 4
axes4 = fig.add_subplot(2, 2, 4)

plt.show()
```

```python
#run all the plotting code together, same as above
fig = plt.figure()
axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)

#add a plot to each of the axes
axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')

#add a title to each subplot
axes1.set_title("Dataset 1")
axes2.set_title("Dataset 2")
axes3.set_title("Dataset 3")
axes4.set_title("Dataset 4")

#add a title to the entire figure
fig.suptitle("Anscombe Data")

#use a tight layout so the plots don't overlap
fig.set_tight_layout(True)

plt.show()
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
#histogram (univariate distribution)
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
axes1.hist(data=tips, x='total_bill', bins=10)
axes1.set_title("Histogram of Total Bill")
axes1.set_xlabel("Frequency")
axes1.set_ylabel("Total Bill")
plt.show()
```

```python
#scatter plot (bivariate distribution)

#create the figure
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)

#add the scatter plot
axes1.scatter(data=tips, x='total_bill', y='tip')

#add labels
axes1.set_title("Scatterplot of Total Bill vs Tip")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tip")
plt.show()
```

```python
#boxplot

#create the figure
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

#add the boxplot
axes1.boxplot(
    #first argument of box plot is the data
    #since we are plotting multiple pieces of data
    #we have to put them in a list
    x = [
        tips.loc[tips["sex"] == "Female", "tip"],
        tips.loc[tips["sex"] == "Male", "tip"]
    ],
    #we can then pass in an optional labels parameter
    #to label the data
    labels = ["Female", "Male"],
)

#add labels
axes1.set_title("Boxplot of Tips by gender")
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
plt.show()
```

```python
#multivariate data
#assign color values
colors = {
    "Female": "#f1a340",
    "Male": "#998fc1",
}

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)

axes1.scatter(
    data=tips,
    x="total_bill",
    y="tip",
    #set the size of the dots based on party size
    s=tips["size"] ** 2 * 10,
    #set the color for the sex using the color dictionary
    c=tips["sex"].map(colors),
    #set the alpha so points are transparent
    alpha=0.6
)

#label the axes
axes1.set_title("Colored by Sex and Sized by Size")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tip")

scatter_plot.suptitle("Total Bill vs Tip")
plt.show()
```

```python
#seaborn
#set the default seaborn context optimized for paper print
#the default is notebook

sns.set_context("paper")
```

```python
#univariate
#histogram
#the subplots function is a shortcut for creating separate figure objects and adding individual subplots or axes to the figure
hist, ax = plt.subplots()

#use seaborn to draw a histogram into the axes
sns.histplot(data = tips, x = "total_bill", ax = ax)

#use matplotlib notation to set the title
ax.set_title("Total Bill Histogram")

#use matplotlib to show the figure
plt.show()
```

```python
#density plot
den, ax = plt.subplots()

sns.kdeplot(data = tips, x = "total_bill", ax = ax)

ax.set_title("Total Bill Density Plot")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Unit Probability")

plt.show()
```

```python
#rug plot
rug, ax = plt.subplots()

#plot 2 things into the axes we created
sns.rugplot(data = tips, x = "total_bill", ax = ax)
sns.histplot(data = tips, x = "total_bill", ax = ax)
ax.set_title("Rug Plot and Histogram of Total Bill")
ax.set_xlabel("Total Bill")

plt.show()
```

```python
#distribution plot
#the facetgrid object creates the figure and axes for us
fig = sns.displot(data = tips, x = "total_bill", kde = True, rug = True)

fig.set_axis_labels(x_var = "Total Bill", y_var = "Count") 
fig.figure.suptitle("Distribution of Total Bill")

plt.show()
```

```python
count, ax = plt.subplots()

#we can use the viridis palette to help distinguish the colors
sns.countplot(data = tips, x = "day", palette = "viridis", ax = ax)

ax.set_title("Count of Days")
ax.set_xlabel("Day")
ax.set_ylabel("Count")

plt.show()
```

```python
#bivariate data
#scatter plot
scatter, ax = plt.subplots()

sns.scatterplot(data = tips, x = "total_bill", y = "tip", ax = ax)

ax.set_title("Scatter Plot of Total Bill vs Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")

plt.show()
```

```python
fig = sns.lmplot(data = tips, x = "total_bill", y = "tip")

plt.show()
```

```python
joint = sns.jointplot(data = tips, x = "total_bill", y = "tip")
joint.set_axis_labels(xlabel = "Total Bill", ylabel = "Tip")
joint.figure.suptitle("Joint Plot of Total Bill and Tip", y = 1.02)
plt.show()
```

```python
#hexbin plot
hexbin = sns.jointplot(data = tips, x = "total_bill", y = "tip", kind = "hex")
hexbin.set_axis_labels(xlabel = "Total Bill", ylabel = "Tip")
hexbin.fig.suptitle("Hexbin Plot of Total Bill and Tip", y = 1.02)
plt.show()
```

```python
#2d density plot
kde, ax = plt.subplots()
sns.kdeplot(data = tips, x = "total_bill", y = "tip", shade = True, ax = ax)
ax.set_title("Kernel Density Plot of Total Bill and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
plt.show()
```

```python
kde2d = sns.jointplot(data = tips, x = "total_bill", y = "tip", kind = "kde")
kde2d.set_axis_labels(xlabel = "Total Bill", ylabel = "Tip")
kde2d.fig.suptitle("2D Kernel Density Plot of Total Bill and Tip", y = 1.02)
plt.show()
```

```python
#bar plot
import numpy as np

bar, ax = plt.subplots()
sns.barplot(data = tips, x = "day", y = "total_bill", estimator = np.mean, ax = ax)
ax.set_title("Average Total Bill by Day")
ax.set_xlabel("Time of Day")
ax.set_ylabel("Average Total Bill")
plt.show()
```

```python
#box plot
box, ax = plt.subplots()
sns.boxplot(data = tips, x = "day", y = "total_bill", ax = ax)
ax.set_title("Box Plot of Total Bill by Day")
ax.set_xlabel("Time of Day")
ax.set_ylabel("Total Bill")
plt.show()
```

```python
violin, ax = plt.subplots()
sns.violinplot(data = tips, x = "day", y = "total_bill", ax = ax)
ax.set_title("Violin Plot of Total Bill by Time of Day")
ax.set_xlabel("Time of Day")
ax.set_ylabel("Total Bill")
plt.show()
```

```python
box_violin, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
sns.boxplot(data = tips, x = "time", y = "total_bill", ax = ax1)
sns.violinplot(data = tips, x = "time", y = "total_bill", ax = ax2)
ax1.set_title("Box Plot")
ax1.set_xlabel("Time of Day")
ax1.set_ylabel("Total Bill")
ax2.set_title("Violin Plot")
ax2.set_xlabel("Time of Day")
ax2.set_ylabel("Total Bill")
box_violin.suptitle("Comparison of Box Plot and Violin Plot")
box_violin.set_tight_layout(True)
plt.show()
```

```python
#pairwise relationships
pair_grid = sns.PairGrid(tips, diag_sharey = False)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.histplot)
plt.show()
```

```python
colors = {"Yes": "#77AADD", "No": "#FFA07A"}
facet2 = sns.relplot(data = tips, x = "total_bill", y = "tip", hue = "smoker", style = "sex", kind = "scatter", col = "day", row = "time",  palette = colors, height = 1.7)
facet2.set_titles(row_template = "{row_name}", col_template = "{col_name}")
sns.move_legend(facet2, loc = "lower center", bbox_to_anchor = (0.5, 0.1), ncol = 2, title = None, frameon = False)
facet2.figure.set_tight_layout(True)
plt.show()
```

```python
seaborn_styles = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

fig = plt.figure()
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(data = tips, x = "time", y = "total_bill", ax = ax)
        violin.set_title(style)
fig.set_tight_layout(True)
plt.show()
```
